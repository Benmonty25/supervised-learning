```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('Resources/2019loans.csv'))
test_df = pd.read_csv(Path('Resources/2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training data
X_train = train_df.drop(['loan_status','index','Unnamed: 0'],axis=1)
Y_train = train_df['loan_status']
X_train
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X_test = test_df.drop(['loan_status','index','Unnamed: 0'],axis=1)
Y_test = test_df['loan_status']

X_test
```

```python
# add missing dummy variables to testing set
X_train_dummies =pd.get_dummies(X_train, drop_first=True)
Y_train = Y_train.replace([{'low_risk': 0,'high_risk':1}])
X_test_dummies = pd.get_dummies(X_test,drop_first= True)
Y_test = Y_test.replace([{'low_risk': 0,'high_risk':1}])
X_test_dummies
```

```python
X_test_dummies['debt_settlement_flag_Y']=0
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()


classifier.fit(X_train_dummies, Y_train)

training_score = classifier.score(X_train_dummies, Y_train)
testing_score = classifier.score(X_test_dummies, Y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(random_state=1, n_estimators=100).fit(X_train_dummies, Y_train)
print(f'Training Score: {clf.score(X_train_dummies, Y_train)}')
print(f'Testing Score: {clf.score(X_test_dummies, Y_test)}')
```

```python
# Scale the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(X_train_dummies)
X_train_scaled = scaler.transform(X_train_dummies)
X_test_scaled = scaler.transform(X_test_dummies)
```

```python
# Train the Logistic Regression model on the scaled data and print the model score

classifier = LogisticRegression().fit(X_train_scaled, Y_train)

training_score = classifier.score(X_train_scaled, Y_train)
testing_score = classifier.score(X_test_scaled, Y_test)

### END SOLUTION 

print(f"Training Score: {training_score}")
print(f"Testing Score: {testing_score}")
```

```python
# Train a Random Forest Classifier model on the scaled data and print the model score
clf = RandomForestClassifier(random_state=1, n_estimators=100).fit(X_train_scaled, Y_train)
print(f'Training Score: {clf.score(X_train_scaled, Y_train)}')
print(f'Testing Score: {clf.score(X_test_scaled, Y_test)}')
```
